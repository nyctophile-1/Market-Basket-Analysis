## Frequent-Free Items

```python
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv("ecommerceData-Revised.csv",index_col=0)
df.head()
```

```python
df.describe()
```

## Here, we see that Minimum Unit Price is 0 for some

```python
df.info()
```

```python
df[df['UnitPrice']==0]
```

```python
df2 = df[df['UnitPrice']==0]
```

```python
baskets = df2.groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo')
```

```python
baskets
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1
basket_sets = baskets.applymap(encode_units)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support = 0.0001, use_colnames = True)
rules = association_rules(frequent_itemsets, metric='confidence',min_threshold=1)
```
